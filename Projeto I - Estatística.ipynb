**3° Tarefa**

**a)** Para fazer um ranking com os acidentes por estado, pode-se usar o comando value_counts(). Antes, irei filtrar as informações que eu quero e juntar os conjuntos de dados de todos os anos. Além do value_counts, usei o sort_values para organizar os dados.

Filtrei as informações de uf e tipo_acidente do dataframe original para criar o ranking.

```python
ranking2020 = acidentes2020[['uf', 'tipo_acidente']]
ranking2019 = acidentes2019[['uf', 'tipo_acidente']]
ranking2018 = acidentes2018[['uf', 'tipo_acidente']]
```

Depois, com o value_counts, e sort_values, organizei os dados do maior para o menor, mostrando os estados com mais acidentes e menos acidentes de cada ano.

```python
data20=ranking2020.value_counts().sort_values(ascending=False)
data19=ranking2019.value_counts().sort_values(ascending=False)
data18=ranking2018.value_counts().sort_values(ascending=False)
```

Depois, joguei os dados em outro dataframe para fazer a concatenação. Além disso, resetei os indices com reset_index(), renomeei a coluna das quantidades de acidentes e criei uma nova coluna com os respectivos anos. Isso é importante para fazer a apresentação dos dados de forma correta e também realizar o multindex.

```python
df20=pd.DataFrame(data=data20).reset_index().rename(columns={0: 'quantidade'}).assign(ano=2020)
df19=pd.DataFrame(data=data19).reset_index().rename(columns={0: 'quantidade'}).assign(ano=2019)
df18=pd.DataFrame(data=data18).reset_index().rename(columns={0: 'quantidade'}).assign(ano=2018)
```

Por fim, setei os indices para fazer o multindex e criei um dataframe final concatenando os dataframes de 2020, 2019 e 2018.

```python
df20.set_index(['ano', 'uf'], inplace=True)
df19.set_index(['ano', 'uf'], inplace=True)
df18.set_index(['ano', 'uf'], inplace=True)
```

```python
ranking = pd.concat([df20, df19, df18], axis=0)
ranking
```

```python
ranking.to_csv('ranking3a.csv')
```

**b)** Para o DataFrame dos dias da semana, pode-se seguir a mesma lógica da questão anterior. A diferença é que, nesse caso, não é necessário usar o sort_values() para fazer a ordenação. Além disso, na função value_counts, usei o argumento sort=False pois não havia a necessidade de organizar os itens em ranking. Exceto isso, a ordem de comandos usados manteve-se

```python
sem20=semana2020.value_counts(sort=False)
sem19=semana2019.value_counts(sort=False)
sem18=semana2018.value_counts(sort=False)
```

```python
dfs20=pd.DataFrame(data=sem20).reset_index().rename(columns={0: 'quantidade'}).assign(ano=2020)
dfs19=pd.DataFrame(data=sem19).reset_index().rename(columns={0: 'quantidade'}).assign(ano=2019)
dfs18=pd.DataFrame(data=sem18).reset_index().rename(columns={0: 'quantidade'}).assign(ano=2018)
```

```python
dfs20.set_index(['ano', 'dia_semana'], inplace=True)
dfs19.set_index(['ano', 'dia_semana'], inplace=True)
dfs18.set_index(['ano', 'dia_semana'], inplace=True)
```

```python
dfsemana = pd.concat([dfs20, dfs19, dfs18], axis=0)
```
