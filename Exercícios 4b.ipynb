**Questão 2:**

Primeiro vou importar os arquivos .csv com os quais irei trabalhar;

Após a importação, irei tratar os dataframes, excluindo dados desnecessários e reorganizando o que for preciso;

Após isso, irei juntar os dataframes em um só e salvar em json;

JSON significa JavaScript Object Notation e é formato leve de troca de informações/dados entre sistemas. Dentre as vantagens do JSON em relação ao CSV temos:

O JSON mantém estruturas hierárquicas, a importação de arquivos JSON costuma dar menos problemas do que em CSV, O JSON dá suporte a listas de objetos, entre outras vantagens.

```python
gasolina2019 = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/vdpb/gasolina-aviacao/vendas-gasolina-aviacao-municipio-2019.csv', sep=';', encoding='utf8')
```

```python
gasolina2017 = pd.read_csv('http://www.anp.gov.br/arquivos/dadosabertos/vc/vendas-gasolinaaviacao-municipio-2017.csv', sep=';', encoding='latin1')
```
