**Questão 1:**

Uma lista é um array que pode conter elementos de vários tipos e esses elementos são acessados através das posições, ou índices, dentro dela.

Um dicionário também pode conter elementos de vários tipos, mas usa, para cada elemento, uma chave correspondente a ele.

Um array numpy possui elementos do mesmo tipo, mas, como a lista, cada um é associado a um índice.

Series é uma estrutura mutável, heterogêneas, vetorizaveis e também indexadas, tanto por posições, como também por índices específicos (ou chaves)

```python
def mudacolunas(df, colu1, colu2):
    colunas = df.columns.tolist()
    i1, i2 = colunas.index(colu1), colunas.index(colu2)
    colunas[i2], colunas[i1] = colunas[i1], colunas[i2]
    return df[colunas]

outdf = mudacolunas(df, 'a', 'c')
```
